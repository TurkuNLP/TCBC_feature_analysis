```python
#Imports
import json
```

```python
#Constants
KEYLISTS = "Keylists.jsonl"
```

```python
keylists = []
with open(KEYLISTS, 'r') as f:
    for line in f:
        keylists.append(json.loads(line))
```

```python
def checkDuplicates(keys: dict):
    trainkeys = keys['train_keys']
    testkeys  = keys['test_keys']+keys['eval_keys']
    for key in trainkeys:
        if key in testkeys:
            return False
    return True

def checkGenres(keys: dict, correct_amounts: dict):
    trainkeys = keys['train_keys']
    amounts = {1:0, 2:0, 3:0}
    for key in trainkeys:
        genre = int(key[-1])
        amounts[genre] = amounts[genre] + 1
    for key in correct_amounts:
        if abs(correct_amounts[key]-amounts[key]) > 1:
            return False
    return True

```

```python
print(checkGenres(keylists[78], {1:158, 2:32, 3:20}))
```

```python
for l in keylists:
    if not checkDuplicates(l):
        print('Key present in both Train and Test!!!!!')
    if not checkGenres(l, {1:158, 2:32, 3:20}):
        print('Genres not properly represented in Train/Test!!!!')
```
